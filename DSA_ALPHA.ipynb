Write a program to rotate an array by k positions using ( TWO POINTERS APPROACH )

```python
def rotate_array(nums, k):
    k = k % len(nums)  # Handle cases where k is greater than the length of the array
    left, right = 0, len(nums) - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1
    left, right = 0, k - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1
    left, right = k, len(nums) - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1
nums = [1, 2, 3, 4, 5]
k = 2
rotate_array(nums, k)
print("Array after rotation: ", nums)  # Output: [4, 5, 1, 2, 3]
```

LEETCODE PROBLEM ON ARRAYS ( CONTAINER WITH MOST WATER )

```python
def maxArea(height):
    left, right = 0, len(height) - 1
    max_area = 0
    while left < right:
        width = right - left
        current_area = min(height[left], height[right]) * width
        max_area = max(max_area, current_area)
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    return max_area
heights = [1, 8, 6, 2, 5, 4, 8, 3, 7]
print("Maximum area: ", maxArea(heights))  # Output: 49
```

LEETCODE PROBLEM ON ARRAYS ( PRODUCT OF ARRAY EXCEPT SELF )

```python
def productExceptSelf(nums):
    length = len(nums)
    answer = [0] * length
    answer[0] = 1
    for i in range(1, length):
        answer[i] = nums[i - 1] * answer[i - 1]
    R = 1
    for i in range(length - 1, -1, -1):
        answer[i] = answer[i] * R
        R *= nums[i]
    return answer
nums = [1, 2, 3, 4]
print("Product of array except self: ", productExceptSelf(nums))  # Output: [24, 12, 8, 6]
```

LEETCODE PROBLEM ON ARRAYS ( 3SUM )

```python
def threeSum(nums):
    nums.sort()
    result = []
    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        left, right = i + 1, len(nums) - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                result.append([nums[i], nums[left], nums[right]])
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
    return result
nums = [-1, 0, 1, 2, -1, -4]
print("Triplets that sum to zero: ", threeSum(nums))  # Output: [[-1, -1, 2], [-1, 0, 1]]
```

 PROBLEM: Given a list of integers, write a program to find the longest consecutive sequence of integers in the list and return its length.       

```python
def longest_consecutive(nums):
    if not nums:
        return 0
    num_set = set(nums)
    longest_streak = 1
    for num in num_set:
        if num - 1 not in num_set:
            current_num = num
            current_streak = 1
            while current_num + 1 in num_set:
                current_num += 1
                current_streak += 1
            longest_streak = max(longest_streak, current_streak)
    return longest_streak
nums = [100, 4, 200, 1, 3, 2]
print("Length of the longest consecutive sequence: ", longest_consecutive(nums))  # Output: 4 (The sequence is [1, 2, 3, 4])
```

example program to understand the SLIDING WINDOW TECHNIQUE

```python
def sliding_window(nums, k):
    window_sum = sum(nums[:k])
    max_sum = window_sum
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]
        max_sum = max(max_sum, window_sum)
    return max_sum
nums = [2,1,5,1,3,2]
k=3
print("Maximum sum of a subarray of size k: ", sliding_window(nums, k))  # Output: 9 (The subarray is [5, 1, 3])
```

```python
def lengthOfLongestSubstring(s):
    char_index = {}
    longest = 0
    start = 0
    for i, char in enumerate(s):
        if char in char_index and char_index[char] >= start:
            start = char_index[char] + 1
        char_index[char] = i
        longest = max(longest, i - start + 1)
    return longest
string = "abcabcbb"
print("Length of the longest substring without repeating characters: ", lengthOfLongestSubstring(string))
# Output: 3 (The substring is "abc")
```

LEETCODE PROGRAM OF FIRST OCCURANCE IN SORTED ARRAY

```python
class Solution:
    def firstOccurrence(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums) - 1
        result = -1  # Default if target not found

        while left <= right:
            mid = left + (right - left) // 2

            if nums[mid] == target:
                result = mid   # Store the index
                right = mid - 1  # Keep searching on the left side for first occurrence
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return result
```

The first occurance program is used to find the first position (index) of a target elements in a sorted array when duplicate values exist.

```python
def first_occurrence(arr, target):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            result = mid
            high = mid - 1   # Move left for first occurrence
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return result


arr = [1, 2, 2, 2, 3, 4, 5]
target = 2

print(first_occurrence(arr, target))
```

COUNT THE NUMBER OF EVEN AND ODD NUMBERS IN THE ARRAY, SIMPLE APPROACH TRAVERSAL

```python
def count_even_odd(arr):
    even = 0
    odd = 0

    for num in arr:
        if num % 2 == 0:
            even += 1
        else:
            odd += 1

    return even, odd


arr = [1, 2, 3, 4, 5, 6, 7, 8]

even, odd = count_even_odd(arr)

print("Even count:", even)
print("Odd count:", odd)
```

SEARCHING AN ELEMENT IN THE SORTED ARRAY (BINARY SEARCH APPROACH)

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1


arr = [2, 4, 6, 8, 10, 12, 14]
target = 10

print(binary_search(arr, target))
```

LOWER BOUND MEANS FINDING THE FIRST INDEX WHERE THE ELEMENTS IS GRATER THAN OR EQUAL TO THE TARGET IN SORTED ARRAY

```python
def lower_bound(arr, target):
    low = 0
    high = len(arr) - 1
    ans = len(arr)

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] >= target:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1

    return ans


arr = [1, 2, 4, 4, 6, 8]
target = 4

print(lower_bound(arr, target))
```

MAXIMUM SUM OF SUB ARRAY USING SLIDING WINDOW CONCEPT

```python
def max_subarray_sum(arr, k):
    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i-k]
        max_sum = max(max_sum, window_sum)

    return max_sum


arr = [2, 1, 5, 1, 3, 2]
k = 3

print(max_subarray_sum(arr, k))
```

SLIDING WINDOW --> SUBARRAY/SUBSTRING PROBLEMS EXAMPLE:

```python
def longest_substring(s):
    left = 0
    max_len = 0
    seen = set()

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1

        seen.add(s[right])
        max_len = max(max_len, right - left + 1)

    return max_len


s = "abcabcbb"

print(longest_substring(s))
```

AGAIN SLIDING WINDOW EXAMPLE:

```python
def max_average(arr, k):
    window_sum = sum(arr[:k])
    max_sum = window_sum

    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i-k]
        max_sum = max(max_sum, window_sum)

    return max_sum / k


arr = [1, 12, -5, -6, 50, 3]
k = 4

print(max_average(arr, k))
```

FIND THE NO OF VOWELS IN A SUBSTRING OF SIZE K

```python
def max_vowels(s, k):
    vowels = "aeiou"

    count = sum(1 for ch in s[:k] if ch in vowels)
    max_count = count

    for i in range(k, len(s)):
        if s[i-k] in vowels:
            count -= 1
        if s[i] in vowels:
            count += 1

        max_count = max(max_count, count)

    return max_count


s = "abciiidef"
k = 3

print(max_vowels(s, k))
```

TWO POINTERS --> PAIR SUM, REMOVE DUPLICATES EXAMPLE:

```python
def remove_duplicates(arr):
    if len(arr) == 0:
        return 0

    i = 0

    for j in range(1, len(arr)):
        if arr[i] != arr[j]:
            i += 1
            arr[i] = arr[j]

    return i + 1


arr = [1, 1, 2, 2, 3, 4, 4, 5]

length = remove_duplicates(arr)

print("Length:", length)
print("Array:", arr[:length])
```

CONTAINER WITH MOST WATER APPROACH : TWO POINTER

```python
def max_area(height):
    left = 0
    right = len(height) - 1
    max_water = 0

    while left < right:
        width = right - left
        area = min(height[left], height[right]) * width

        max_water = max(max_water, area)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_water


height = [1,8,6,2,5,4,8,3,7]

print(max_area(height))
```

MERGE SORTED ARRAY APPROACH: TWO POINTERS

```python
def merge_sorted(arr1, arr2):
    i = 0
    j = 0
    result = []

    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1

    while i < len(arr1):
        result.append(arr1[i])
        i += 1

    while j < len(arr2):
        result.append(arr2[j])
        j += 1

    return result


arr1 = [1, 3, 5]
arr2 = [2, 4, 6]

print(merge_sorted(arr1, arr2))
```

SQUARES OF SORTED ARRAY APPROACH : TWO POINTERS

```python
def sorted_squares(arr):
    left = 0
    right = len(arr) - 1
    result = [0] * len(arr)

    for i in range(len(arr)-1, -1, -1):
        if abs(arr[left]) > abs(arr[right]):
            result[i] = arr[left] ** 2
            left += 1
        else:
            result[i] = arr[right] ** 2
            right -= 1

    return result


arr = [-7, -3, 2, 3, 11]

print(sorted_squares(arr))
```

MINMUM SIZE SUBARRAY SUM APPROACH: TWO POINTERS

```python
def min_subarray_len(target, arr):
    left = 0
    window_sum = 0
    min_len = float('inf')

    for right in range(len(arr)):
        window_sum += arr[right]

        while window_sum >= target:
            min_len = min(min_len, right - left + 1)
            window_sum -= arr[left]
            left += 1

    return 0 if min_len == float('inf') else min_len


target = 7
arr = [2, 3, 1, 2, 4, 3]

print(min_subarray_len(target, arr))
```

SORT COLORS (DUTCH NATIONAL FLAG PROGRAM) APPROACH: THREE POINTERS

```python
def sort_colors(nums):
    low = 0
    mid = 0
    high = len(nums) - 1

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1

        elif nums[mid] == 1:
            mid += 1

        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1

    return nums


nums = [2, 0, 2, 1, 1, 0]

print(sort_colors(nums))
```

LINKED LIST CYCLE DETECTION APPROACH: SLOW & FAST POINTERS

```python
class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None


def has_cycle(head):
    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True

    return False


# Example
head = ListNode(1)
second = ListNode(2)
third = ListNode(3)
fourth = ListNode(4)

head.next = second
second.next = third
third.next = fourth
fourth.next = second   # Creates cycle

print(has_cycle(head))
```

SEARCH INSERT POSITION - FIND THE INDEX OF THE TARGET. IF NOT FOUND RETURN THE POSITION WHERE IT SHOULD BE INSERTED

```python
def search_insert(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return low


arr = [1, 3, 5, 6]
target = 2

print(search_insert(arr, target))
```

FLOOR AND CEIL IN SORTED ARRAY - APPROACH : BINARY SEARCH , FLOOR(GRATEST ELEMENT <= TARGET), CEIL(SMALLEST ELEMENT <= TARGET)

```python
def floor_ceil(arr, target):
    low = 0
    high = len(arr) - 1
    floor = -1
    ceil = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return arr[mid], arr[mid]

        elif arr[mid] < target:
            floor = arr[mid]
            low = mid + 1

        else:
            ceil = arr[mid]
            high = mid - 1

    return floor, ceil


arr = [1, 2, 4, 6, 8, 10]
target = 5

f, c = floor_ceil(arr, target)

print("Floor:", f)
print("Ceil:", c)
```

FIND PEAK ELEMENT APPROACH: BINARY SEARCH

```python
def find_peak(arr):
    low = 0
    high = len(arr) - 1

    while low < high:
        mid = (low + high) // 2

        if arr[mid] < arr[mid + 1]:
            low = mid + 1
        else:
            high = mid

    return low


arr = [1, 2, 1, 3, 5, 6, 4]

print(find_peak(arr))
```

SEARCH IN ROTATED SORT ARRAY APPROACH: BINARY SEARCH

```python
def search_rotated(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid

        # Left half is sorted
        if arr[low] <= arr[mid]:
            if arr[low] <= target < arr[mid]:
                high = mid - 1
            else:
                low = mid + 1

        # Right half is sorted
        else:
            if arr[mid] < target <= arr[high]:
                low = mid + 1
            else:
                high = mid - 1

    return -1


arr = [4,5,6,7,0,1,2]
target = 0

print(search_rotated(arr, target))
```

FIND THE MINIMUM IN ROTATED SORTED ARRAY APPROACH: BINARY SEARCH

```python
def find_min(arr):
    low = 0
    high = len(arr) - 1

    while low < high:
        mid = (low + high) // 2

        if arr[mid] > arr[high]:
            low = mid + 1
        else:
            high = mid

    return arr[low]


arr = [4,5,6,7,0,1,2]

print(find_min(arr))
```

SINGLE ELEMENT IN SORTED ARRAY APPROACH: BINARY SEARCH

```python
def single_non_duplicate(arr):
    low = 0
    high = len(arr) - 1

    while low < high:
        mid = (low + high) // 2

        # Make mid even
        if mid % 2 == 1:
            mid -= 1

        if arr[mid] == arr[mid + 1]:
            low = mid + 2
        else:
            high = mid

    return arr[low]


arr = [1,1,2,2,3,4,4,5,5]

print(single_non_duplicate(arr))
```
